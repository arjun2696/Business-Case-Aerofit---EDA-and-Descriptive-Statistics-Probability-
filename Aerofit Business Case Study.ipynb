```python
#As induced from the plot above , we can observe that there many outliers in the income column

#We will find the number of outliers present in the column
```

```python
income_25 = df["Income"].quantile(0.25)
income_50 = df["Income"].quantile(0.5)
income_75 =df["Income"].quantile(0.75)
IQR_income  = income_75-income_25
outlier_income = income_75+(1.5*(IQR_income))
len(df[df["Income"]>outlier_income])/len(df)*100
```

```python
#This shows that 10 percent of the values are outliers in income
#Removing the outlier will not make sense as the data is very small
```

```python
#Converting Age into 4 different categories 
```

```python
def ChangeAge(x):
    if x >=18 and x<26:
        return "Young Age"
    if x>=26 and x<34:
        return "Mid Age"
    if x>=34 and x<42:
        return "High Mid Age"
    else:
        return "Senior Age"
```

```python
df2["Age Groups"] = df2["Age"].apply(ChangeAge)
df2
```

```python
sns.countplot(x = "Age Groups" ,  hue = "Product" , data = df2)
```

```python
pd.crosstab(index = df2["Product"] , columns=df2["Age Groups"], normalize="index")
```

```python
#There is low percentage of people at senior age and high mid age going for KP781
```

```python
lowincome=df2["Income"].quantile(0.33)
midincome=df2["Income"].quantile(0.66)
```

```python
def ChangeIncome(x):
    if x <= lowincome:
        return "Low"
    if x > lowincome and x<=midincome:
        return "Mid"
    else:
        return "High"
    
```

```python
df2["Income Group"] = df2["Income"].apply(ChangeIncome)
```

```python
df2["Income Group"].value_counts(normalize =True)
```

```python
def ChangeMiles(x):
    if x < 100:
        return "Low"
    if x >= 100 and x <200:
        return "Mid"
    else:
        return "High"
```
